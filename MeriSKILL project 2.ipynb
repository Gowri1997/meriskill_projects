## Data PreprocessingÂ¶

#### Feature engineering

```python
plt.figure(figsize=(10,7))
sn.heatmap(df.corr(),annot=True,fmt=".2%",cmap='coolwarm')
```

#### Outlier Removal

```python
avg = df['BloodPressure'].mean()
df['BloodPressure'] = df['BloodPressure'].apply(lambda v: avg if v==0 else v)
```

```python
plt.figure(figsize=(8,5))
sn.distplot(df['BloodPressure'], bins = 30,color='orange')
plt.title('Distribution of BloodPressure')
```

```python
bmi_avg = df['BMI'].mean()
df['BMI'] = df['BMI'].apply(lambda v: bmi_avg if v==0 else v)
```

```python
plt.figure(figsize=(8,5))
sn.displot(df,x='BMI',color='black',kind="kde")
plt.title('Distribution of BMI')
```

As seen earlier there are amount which varies largely from rest of amount.It may not me outlier in real case but for better modelling we need to remove it.In above variable i replaced outlier value with  mean.
Now lets use z-score technique to see outlier in other feature variables.

```python
df['z_score_bp']=(df.BloodPressure-df.BloodPressure.mean())/df.BloodPressure.std()
df['z_score_Glucose']=(df.Glucose-df.Glucose.mean())/df.Glucose.std()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(columns=['bmi_lab','Age_group','z_score_bp','z_score_Glucose'],inplace=True)
df.head()
```

## Split Data

```python
X=df.drop(columns='Outcome')
Y=df['Outcome']
```

```python
X.head()
```

```python
Y.head()
```

```python
X.shape,Y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Model building

## Logistic regression

```python
from sklearn.linear_model import LogisticRegression
m_1=LogisticRegression()
```

```python
m_1.fit(x_train,y_train)
```

```python
y_predicted_l=m_1.predict(x_test)
```

```python
m_1.score(x_test,y_test)
```

Evaluate the models performance using matrics like accuracy,confution matric and classification report

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
accuracy_l=accuracy_score(y_test,y_predicted_l)
accuracy_l
```

```python
cm_l=confusion_matrix(y_test,y_predicted_l)
print(cm_l)
```

```python
class_report_l=classification_report(y_test,y_predicted_l)
print(class_report_l)
```
